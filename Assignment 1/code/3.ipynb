```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
Xa = np.linspace(-3, 3, 100)
Ya = np.full(100,1/6)

Xb = np.linspace(-8, 8, 200)
Yb = np.exp(-(Xb*Xb)/6)/np.sqrt(6*np.pi)

plt.plot(Xa, Ya, color='blue')
plt.plot(Xb, Yb, color='orange')
plt.plot([0,0], [0,1/np.sqrt(6*np.pi)], color='green', linestyle = '--')

plt.plot([-3,-3], [0,1/6], color='blue')
plt.plot([3,3], [0,1/6], color='blue')
plt.plot([-8,-3], [0,0], color='blue')
plt.plot([3,8], [0,0], color='blue')

plt.legend(["Uniform density", "Normal density", "Mean"])
plt.xlabel("x")
plt.ylabel("p(x)", rotation = 0)
plt.xlim([-6, 6])
plt.ylim([0, 0.25])
plt.savefig("Probability distributions.png")
plt.show()
```
